```python
from matplotlib import pyplot as plt

x=[5,2,7]
y=[2,16,4]
plt.plot(x,y)
plt.title("Info")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May','June']
energy_consuption=[1200,1300,1100,1400,1500,1000]
plt.plot(months,energy_consuption,marker='*',color='r',linestyle='--')
plt.title("Energy Consuption over 6 months")
plt.xlabel("Month")
plt.ylabel("Energy consuption in MWh")
plt.show()
```

```python
energy_sources=['Solar','Wind','Hydropower','Biomass']
energy_values=[1200,3400,2900,3500]
plt.bar(energy_sources,energy_values,color='orange')
plt.title("Energy consuption by renewable sources of energy")
plt.xlabel("Energy Sources")
plt.ylabel("Energy consuption (MWh)")
plt.show()
```

```python
plt.pie(energy_values,labels=energy_sources,autopct='%1.1f%%',colors=['gold','orange','lightgreen','coral'])
plt.title("Energy consuption share by source")
plt.show()
```

```python
# Sample data for energy consumption and carbon emissions
carbon_emissions = [200, 500, 450, 300]  # in kg CO2

# Create a scatter plot
plt.scatter(energy_values, carbon_emissions, color='red')

# Add titles and labels
plt.title('Energy Consumption vs Carbon Emissions')
plt.xlabel('Energy Consumption (MWh)')
plt.ylabel('Carbon Emissions (kg CO2)')
plt.show()
```

```python
# Create a customized bar chart
plt.bar(energy_sources, energy_values, color='cyan', edgecolor='black', linewidth=2)

# Add titles and labels
plt.title('Customized Energy Consumption by Source', fontsize=14, fontweight='bold')
plt.xlabel('Energy Source', fontsize=12)
plt.ylabel('Energy Consumption (MWh)', fontsize=12)

# Adding grid and customizing style
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(['Energy Consumption'], loc='upper left')
plt.show()
```
