```python
from matplotlib import pyplot as plt

x=[5,2,7]
y=[2,16,4]
plt.plot(x,y)
plt.title("Info")
plt.xlabel("X axis")
plt.ylabel("Y axis")
plt.show()
```

```python
months=['Jan','Feb','Mar','Apr','May','June']
energy_consuption=[1200,1300,1100,1400,1500,1000]
plt.plot(months,energy_consuption,marker='*',color='r',linestyle='--')
plt.title("Energy Consuption over 6 months")
plt.xlabel("Month")
plt.ylabel("Energy consuption in MWh")
plt.show()
```

```python
energy_sources=['Solar','Wind','Hydropower','Biomass']
energy_values=[1200,3400,2900,3500]
plt.bar(energy_sources,energy_values,color='orange')
plt.title("Energy consuption by renewable sources of energy")
plt.xlabel("Energy Sources")
plt.ylabel("Energy consuption (MWh)")
plt.show()
```
