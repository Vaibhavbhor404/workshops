```python
total_consuption=np.sum(energy_consuption)
print(f"Total energy consuption :{total_consuption} Mwh")
```

```python
mean_consuption=np.mean(energy_consuption)
print(f"Mean energy consuption :{mean_consuption:.2f}")
```

```python
std_deviation=np.std(energy_consuption)
print(f"Standard deviation of energy consuption :{std_deviation:.2f} Mwh")
```
