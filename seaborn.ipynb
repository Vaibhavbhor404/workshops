```python
import seaborn as sns
sns.set(style='whitegrid')
tips=sns.load_dataset('tips')
numeric_data=tips.select_dtypes(include='number')
sns.heatmap(numeric_data.corr(),annot=True,cmap='rainbow')
numeric_data.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
data={
    "Solar":[1200,1500,1300],
    "Wind":[3400,3600,3200],
    "Hydropower":[2900,3100,2800],
    "Biomass":[2500,2700,2400]
}
df=pd.DataFrame(data)
sns.pairplot(data=df)
plt.show()
```

```python
corr_matrix=df.corr()
corr_matrix
```

```python
corr_matrix=df.corr()
sns.heatmap(corr_matrix,annot=True,cmap='rainbow',linewidths=0.5)
plt.title("Correlation between energy resources")
plt.show()
```

```python
data={
    "Solar":[1200,1500,1300],
    "Wind":[3400,3600,3200],
    "Hydropower":[2900,3100,2800],
    "Biomass":[2500,2700,2400]
}
df=pd.DataFrame(data)
sns.boxplot(data=df)
plt.title("Distribution of energy consuption by sources")
plt.show()
```

```python
sns.violinplot(data=df)
plt.title("Violin plot of energy consuption by sources")
plt.show()
```

```python
energy_values=[100,200,300,400,500]
carbon_emission=[10,20,30,40,50]
df_reg=pd.DataFrame({'Energy Consuption' : energy_values, 'Carbon Emission' : carbon_emission})
sns.regplot(x='Energy Consuption',y='Carbon Emission',data=df_reg)
plt.title("Energy consuption vs carbon Emission")
plt.xlabel("Energy Consuption")
plt.ylabel("Carbon Emission")
plt.show()
```
