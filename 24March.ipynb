```python
import pandas as pd
import numpy as np
data={
    "Energy source":["Solar","wind","Hydropower","Geothermal","Biomass","Nuclear"],
    "Energy Consuption (MWh)":[1200,np.nan,2900,np.nan,2500,3200],
    "Cost (million $ )":[200,400,np.nan,150,250,np.nan]
}
df=pd.DataFrame(data)
# print(df)
# df.head(3)
df.tail(3)
```

```python
cleaned_df=df.dropna()
print(cleaned_df)
```

```python
df.columns
```

```python
df['Energy Consuption (MWh)'].fillna(df['Energy Consuption (MWh)'].mean(),inplace=True)
df['Cost (million $ )'].fillna(df['Cost (million $ )'].mean(),inplace=True)
print(df)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
df[['Energy Consuption (MWh)','Cost (million $ )']]=scaler.fit_transform(df[['Energy Consuption (MWh)','Cost (million $ )']])
print(df)
```

```python
import pandas as pd
energy_encoded_df=pd.get_dummies(df,columns=['Energy source'])
print("\nData after one-hot Encoding Categorical variable")
print(energy_encoded_df)
```

```python
energy_encoded_df['Cost (million $ )']=energy_encoded_df["Energy Consuption (MWh)"]
print("\nData with New Feature (Consuption per $Million) :")
print(energy_encoded_df)
```
